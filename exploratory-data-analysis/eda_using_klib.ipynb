```python
## Importing libraries

import pandas as pd
import numpy as np
import klib
```

```python
## Loading Data

# Data URL from UCI repository
auto_data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'

# List of columns
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
                'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 
                'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 
                'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

# Reading data as pandas dataframe
df = pd.read_csv(auto_data_url, names=column_names)
df = df.replace({'?': np.nan})
```

```python
## Converting columns to Numeric type
numeric_cols = ['normalized-losses', 'engine-size', 'horsepower', 'peak-rpm', 'bore', 'stroke', 'price']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

## klib.describe - functions for visualizing datasets

```python
## Display distribution of categorical column values
klib.cat_plot(df, figsize=(30,25))
```

```python
## Getting correlation among variables
klib.corr_mat(df) # returns a color-encoded correlation matrix
```

```python
## Getting heatmap for correlation among variables
klib.corr_plot(df) # returns a color-encoded heatmap, ideal for correlations
```

```python
## Display only positive correlation [To get only negative correlation use split='neg']
klib.corr_plot(df, split='pos', annot=True)
```

```python
## Displaying correlation with a target variable
klib.corr_plot(df, target='price')
```

```python
## Displaying value distribution of numeric features
klib.dist_plot(df) # returns a distribution plot for every numeric feature
```

```python
## Displaying missing values in each column
klib.missingval_plot(df) # returns a figure containing information about missing values
```

```python
## Saving any plot as a pdf file
plot = klib.corr_plot(df, split='neg')
plot.figure.savefig('figure.pdf')
```
